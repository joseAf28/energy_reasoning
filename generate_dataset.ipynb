```python
import sudoku_dataset as sd
import numpy as np
from tqdm import tqdm 
from torch.utils.data import DataLoader
```

```python

num_samples = 25_000
holes = 40

gen = sd.SudokuGenerator()

pbar = tqdm(total=num_samples)

data = []
for _ in range(num_samples):
    puzzle, sol = gen.make_puzzle(holes=holes)
    data.append((puzzle, sol))
    pbar.update(1)

pbar.close()
```

```python

size_train = int(0.80 * num_samples)
size_test = num_samples - size_train
### store data

data_train = np.array([item for item in data[:size_train]])
data_test = np.array([item for item in data[size_train:]])
```

```python
np.save('dataset_train.npy', data_train)
np.save('dataset_test.npy', data_test)
```

```python

dataset = sd.FileDataset("dataset_train.npy")
dataloader = DataLoader(dataset, batch_size=64, shuffle=True)
```

```python

epoch = 0
for batch in tqdm(dataloader, desc=f"Epoch {epoch+1}/100"):
    print(batch['puzzle'].shape)
```
